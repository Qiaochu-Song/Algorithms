```python
import copy
import numpy as np

class Vertex:
    def __init__(self,key):
        self.idvalue = key
        self.connection = [] 
        
    def addneighbors(self,vertices):    #here 'vertices' is a list of vertex objs
        for i in vertices:
            self.connection.append(i)
    
class Graph:
    def __init__(self):    
        self.idlist = []
        self.nodes = {}      # use dict here
        self.size = 0
        
    def addVertex(self,key):   # key: vertex name, i.e., idvalue
        self.idlist.append(key) 
        self.nodes[key]= Vertex(key)    # use g.nodes[ind] to refer to each vertex
        self.size += 1
        
    def copy(self):
        _g = Graph()
        _g.idlist = self.idlist.copy()
        _g.nodes = copy.deepcopy(self.nodes)
        _g.size = self.size
        return _g
        
    def randEdge(self):
        i = np.random.randint(0,len(self.idlist))
        j = np.random.randint(0,len(self.nodes[self.idlist[i]].connection))
        return self.idlist[i],self.nodes[self.idlist[i]].connection[j].idvalue
    
    def mergenodes(self,key1,key2): 
        #L1 = len(self.nodes[key1].connection)
        #L2 = len(self.nodes[key2].connection)
        i1 = 0
        i2 = 0
        while i2 < len(self.nodes[key2].connection):          # delete ALL self loops (cannot use remove here,since there might be >1 self loops)
            if self.nodes[key2].connection[i2] == self.nodes[key1]:   #to guarantee we will not pick E(key1-key2) again
                self.nodes[key2].connection.pop(i2)   
            else:
                i2+=1
                
        
        while i1 < len(self.nodes[key1].connection): 
            if self.nodes[key1].connection[i1] == self.nodes[key2]:
                self.nodes[key1].connection.pop(i1) 
            else:
                i1+=1  
        
        addconnect = self.nodes[key1].connection + self.nodes[key2].connection 
        self.nodes[key1].connection = addconnect   # add vertex[key2] neighbors (except key1) to V[key1]
        self.idlist.remove(key2)   # delete key2 from idlist(it is ok to use remove here)
        
        for v in self.nodes[key2].connection:
            for t in range(0,len(v.connection)):
                if v.connection[t] == self.nodes[key2]:
                    v.connection[t] = self.nodes[key1] #change all nodes that 'point at key2' to 'point at key1'
                 
        
        self.size = len(self.idlist)
        
        
    def RandContract(self):
        while self.size>2:
            key1,key2 = self.randEdge()
            self.mergenodes(key1,key2)
        return len(self.nodes[self.idlist[0]].connection) 
    
    def minCut(self):
        num = self.size
        for i in range(0,self.size**2):
            gtemp = self.copy()   # in order to keep every input in each loop the same as 'self'
            output = gtemp.RandContract()
            if output < num:
                num = output 
        return num
# SMALL GRAPH TEST
# list1 = [[1,3,4],[2,3,4],[3,1,2,5],[4,1,2],[5,3]]
# g2 = Graph()
# for i in list1:
#     g2.addVertex(i[0])
# del i 
# for i in list1:
#     neighbors = []
#     for j in i[1:]:
#         neighbors.append(g2.nodes[j])
#     g2.nodes[i[0]].addneighbors(neighbors)


# #g2.nodes[1].connection[2].idvalue      # now function 'mergenodes' and RandContract are right
# g2.size

# #g2.minCut()  #something wrong with this function
# _g2 = g2.copy()

# _g2.minCut()
# #_g2.mergenodes(3,5)
# #_g2.nodes[3].connection
                
            
        
```

```python
f = open('KargerMinCut.txt','r')
g = Graph()

for line in f: 
    line = line.rstrip('\t \n')
    line = list(line.split('\t'))
    g.addVertex(line[0])   #convert each line into an empty vertex directly  
    
f.seek(0)

for line2 in f:        # add neighboring vertices to each vertex (by referring each other)
    vertices=[]
    line2 = line2.rstrip('\t \n')
    line2 = list(line2.split('\t'))
    for i in line2[1:]:
        vertices.append(g.nodes[i])
    g.nodes[line2[0]].addneighbors(vertices) #line[0] is np.node

f.close()

#len(g.nodes['200'].connection)
g.size
```
